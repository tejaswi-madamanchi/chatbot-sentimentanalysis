```python
import torch
from transformers import TrainingArguments, Trainer, AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_dataset
import pandas as pd
```

```python
df = pd.read_csv("/content/Dataset-SA.csv")
df = df.rename(columns={"Review": "text", "Sentiment": "label"})
df = df[["text", "label"]]
```

```python
label_mapping = {"positive": 2, "neutral": 1, "negative": 0}
df["label"] = df["label"].map(label_mapping)
```

```python
df.to_csv("processed_dataset.csv", index=False)  # Save for dataset loading
dataset = load_dataset("csv", data_files={"train": "processed_dataset.csv"})
```

```python
model_name = "deepseek-ai/deepseek-coder-7b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)
```

```python
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
num_labels = 3  # Negative, Neutral, Positive
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,  # Optimize for Colab GPU
    per_device_eval_batch_size=16,
    num_train_epochs=1,  # Faster training
    weight_decay=0.01,
    logging_dir="./logs",
    fp16=True,  # Use mixed precision
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
)

```

```python
# ✅ Step 12: Train model
trainer.train()

# ✅ Step 13: Save trained model
model.save_pretrained("deepseek_sentiment_model")
tokenizer.save_pretrained("deepseek_sentiment_model")
```
